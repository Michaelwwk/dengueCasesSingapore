### Combine datasets

```python
combined = pd.concat([dengue_cases_condensed, water_bodies_condensed, parks, hawkers_condensed], axis = 0)
```

```python
combined = combined[~((combined.Type == 'Dengue Cases') & (combined.case_count == 0.0))]
```

```python
combined.loc[combined.Type != 'Dengue Cases', 'case_count'] = 10
```

```python
# Define the types and years
Types = ['Water Bodies', 'Parks', 'Hawkers']
Years = range(2013, 2021)

# Create an empty list to hold the data
data = []

# Loop through the types and years, and append a row to the data list for each combination
for t in Types:
    for y in Years:
        data.append([t, y])
        
# Create the dataframe from the data list
df = pd.DataFrame(data, columns=['Type', 'year'])
```

```python
combined = pd.merge(combined, df, how = 'left', on = 'Type')
```

```python
combined.loc[combined.Type == 'Dengue Cases', 'year_y'] = combined['year_x']
combined.drop(columns = 'year_x', inplace = True)
combined.rename(columns = {'year_y': 'year'}, inplace = True)
combined['year'] = combined.year.astype('Int64')
```

```python
combined['Planning Area'] = combined['Planning Area'].str.lower()
combined['Planning Area'] = combined['Planning Area'].str.title()
```
